```r
i <-  1
mean_71 <- 0
tracker <- 0
while (i <= nrow(dataset)){
    father_dat = dataset[i, 1]
    son_dat = dataset[i, 2]
    if (round(father_dat) == 71){
        mean_71 <- mean_71 + son_dat
        tracker <- tracker + 1
    }
    i <- i + 1
}
```

```r
mean_71/tracker
```
