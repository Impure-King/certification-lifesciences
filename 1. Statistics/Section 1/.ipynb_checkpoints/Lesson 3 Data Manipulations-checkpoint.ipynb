```r
library(dplyr)
```

```r
# Loading the data:
dat <- read.csv("data/femaleMiceWeights.csv")
```

```r
# How to filter the data:
controls <- filter(dat, Diet=="chow")
```

```r
# Selecting a specific data:
controls_dat <- unlist(select(controls, Bodyweight))
```

```r
controls_dat
```

```r
# Creating a pipeline:
filtered_data <- filter(dat, Diet=="chow") %>% select(Bodyweight) %>% unlist()
```

```r
filtered_data
```

The pipeline passes the output of the first line to next line as the first argument. This is a shorthand way to write a nice function and will organize the code.
